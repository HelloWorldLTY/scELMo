```python
# python3
# please install OpenAI SDK first: `pip3 install openai`
from openai import OpenAI

client = OpenAI(api_key="use your api", base_url="https://api.deepseek.com") #paid deepseek as example

response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "Please help me summarize the function of gene: CD79A."},
    ]
)

output = response.choices[0].message.content
```

```python
from FlagEmbedding import FlagModel #open-source embeddings
model = FlagModel('BAAI/bge-small-en-v1.5', 
                  use_fp16=True) # Setting use_fp16 to True speeds up computation with a slight performance degradation
embeddings_1 = model.encode(output)
```
